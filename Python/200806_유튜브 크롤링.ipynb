```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

main_url='https://www.youtube.com/'
keyword='머신러닝'
options=webdriver.ChromeOptions()  # 크롬 브라우저가 뜨는 것 없이 진행
options.add_argument('headless')
web_driver = webdriver.Chrome('chromedriver.exe',chrome_options=options)
web_driver.get(main_url)
```

```python
search_text=web_driver.find_element_by_name('search_query')
search_text.send_keys(keyword)
search_text.send_keys(Keys.RETURN)
```

```python
import time
elem=web_driver.find_element_by_tag_name('body')
for page_down in range(1,50):
    print('page_down=',page_down)
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

```python
html_source=web_driver.page_source
```

```python
from bs4 import BeautifulSoup as bs
soup=bs(html_source,'html.parser')
```

```python
all=soup.find_all('a',{'id':'video-title'})
title=[]
for a in all:
    title.append(a.text.replace('\n',''))
print(title)
```
