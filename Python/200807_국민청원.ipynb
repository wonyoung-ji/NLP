#### 별도 answer 컬럼부여 + 20만건 이상 TRUE / 아니면 FALSE

```python
answer=[]
for a in df['votes']:
    if a >= 200000:
        b=True
        answer.append(b)
    else: 
        b=False
        answer.append(b)
df['answer']=answer
```

```python
# df['answer']=df['votes']>20000
```

#### answer컬럼의 타입을 int로 변환

```python
df.info()
```

```python
df['answer']=df['answer'].astype('int')   # boolean 이었던 타입을 int 타입으로 변환
df.info()
```
