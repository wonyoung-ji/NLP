```python
x=np.linspace(0,2,100)
fig, ax=plt.subplots()
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubric')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out=ax.plot(data1,data2, **param_dict)
    return out
```

```python
data1, data2= np.random.randn(2,100)
fig, ax= plt.subplots(1,1)
my_plotter(ax,data1,data2,{'marker':'x'})
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data1,data2,{'marker':'o'})
```

```python
def my_plotter(ax,data1,data2,param_dict):
    out=ax[0][0].plot(data1,**param_dict)
    out=ax[1][0].plot(data2, **param_dict)
    return out
data1,data2=np.random.randn(2,100)
fig, ax=plt.subplots(2,2)
my_plotter(ax,data1,data2,{'marker':'x'})
```

```python
data={'a':np.arange(50),'c':np.random.randint(0,50,50),'d':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100

plt.scatter('a','b',c='c', s='d',data=data, alpha=0.8)  #c는 컬러, s는 marksize, alpha는 투명도
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.show()
```

```python
names=['group_a','group_b','group_c','group_d']
values=[1,10,100,1000]

plt.figure(figsize=(10,2))

plt.subplot(141) #1x4의 첫 번째
plt.bar(names,values)

plt.subplot(142) #1x4의 두 번째
plt.scatter(names,values)

plt.subplot(143) #1x4의 세 번째
plt.plot(names,values)

plt.subplot(144)
plt.plot(names,values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
mu, sigma=100,15
x=mu+sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='skyblue',alpha=0.4)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,0.025,r'$\mu=100,\ \sigma=15$')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')
tips=sns.load_dataset('tips')  #seaborn 사이트 내 데이터 불러오기(df값으로 저장됨)
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips)

#팔레트: 내가 색깔 지정 안하고, 특정 계열로 바꿀 때
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights,annot=True, fmt='d',linewidths=.5,ax=ax)
```
