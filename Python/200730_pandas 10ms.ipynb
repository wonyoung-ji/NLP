```python
left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})
print(left)
print(right)
```

```python
pd.merge(left,right,how='left',on='key') #left 키를 기준으로:1-4,1-5 / 2-4,2-5
```

```python
pd.merge(left,right,how='right',on='key') #right키를 기준으로: 4-1,4-2 / 5-1,5-2
```
