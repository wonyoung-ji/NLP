```python
#2 합격자

for i in test_score.items():
    name=i[0]
    score=i[1]
    average=sum(score)/len(score)
    if average>=60 and min(score)>=40:
        print(name)
```

```python
#3 과목별 최고점수, 최저점수

df=pd.DataFrame(test_score,index=['국어','영어','수학','과학'])
print('최고점')
print(df.max(axis=1))

print()

print('최저점')
print(df.min(axis=1))
```

```python
#선생님 정답

pass_person=[]
for i in test_score.items():
    name=i[0]
    score=i[1]
    average=sum(score)/len(score)
    print('이름:',name, '평균:',average, '최고점:',max(score), '최저점:', min(score))
    if average>=60 and min(score)>=40:
        pass_person.append(name)
        
print('합격자:',pass_person)
```
