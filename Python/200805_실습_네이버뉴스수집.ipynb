## 네이버 뉴스 수집

```python
import requests
from bs4 import BeautifulSoup as bs

url="https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
html=requests.get(url).text
soup=bs(html,'html.parser')
```

```python
all=soup.find('ul',{'class':'type06_headline'})
dl=all.find_all('dl')    # dl 태그 붙은 걸 다 가져옴
dl
```

```python
for item2 in dl:
    try:
        img = item2.find('dt',{'class':'photo'}).find('img')
        print('img=',img['src'])
        print('-'*40)
    except:
        print('No image')
```

```python
dl=all.find_all('dl')
for idx,item2 in enumerate(dl):
    link = item2.find('dt',{'class':''}).find('a')   ## dt태그인데 class가 없는
    print(link)
    print(idx,'번',link['href'])    
    print(link.text.replace('\t','').replace('\n','')[1:len(link.text)+1])
    print('')
```

```python
dl=all.find_all('dl')
```

```python
for item2 in dl:
    try:
        content=item2.find('dd')
        print(content)
        print(content.text.replace('\t','').replace('\n','').split('…')[0])
    except:
        print('No content')
```
