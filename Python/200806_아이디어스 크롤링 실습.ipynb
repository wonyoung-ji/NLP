```python
import requests
from bs4 import BeautifulSoup as bs

url='https://www.idus.com/'
html=requests.get(url).text
soup=bs(html,'html.parser')
```

```python
a=soup.find_all('ul',{'class':'ui_grid__cols--5'})
title=[]
for item in a:
    try:
        content=item.find('li',{'class':'ui_grid__item'}).find('div',{'class':'ui_card'}).find('div',{'class':'ui_card__info'}).find('a',{'class':'ui_card__title'})
        title.append(content.text)
    except:
        print('none')
title
```

```python
rating=[]
for item in a:
    try:
        content2=item.find('li',{'class':'ui_grid__item'}).find('div',{'class':'ui_card'}).find('div',{'class':'ui_card__rating'})
        rating.append(content2.text.replace('\n',''))  
    except:
        print('none')
rating
```

```python
b=soup.find_all('ul',{'class':'ui_gnb'})
category=[]
for items in b:
    content=items.find('li',{'class':'ui_gnb__menu'})
    category.append(content.text.replace('\n',''))
    
category
```
